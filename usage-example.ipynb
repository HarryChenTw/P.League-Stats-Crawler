### 得到比賽數據

```python
from PStatsScraper.GameStatisticTable import GameStatisticTable
print('支援的賽季 : ', GameStatisticTable.supported_season)
print('可以選擇的比賽類型 : ', GameStatisticTable.supported_gmae_types)
```

```python
game_statistic_table = GameStatisticTable(season='2021-22',game_type='playoffs',game_number='AG1')
table = game_statistic_table.get_table(team_side='home')
table
```

```python
# download table
game_statistic_table.download_table(team_side='home',format='xlsx',output_path='test.xlsx')
```

```python
# aslo provide english column version
game_statistic_table = GameStatisticTable(season='2021-22',game_type='playoffs',game_number='AG1', english_col=True)
table = game_statistic_table.get_table(team_side='home')
table
```
